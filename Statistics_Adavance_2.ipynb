#Theory Questions

1. What is hypothesis testing in statistics?
    - Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It helps to determine whether a claim or assumption about a population is true or false using evidence from data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
   - Null Hypothsis: It states that there is no effect or no difference. It is used as the initial or starting or default assumption.
   - Alternative hypothesis: It states that there is an effect or a difference. This is what you aim to prove.

3. What is the significance level in hypothesis testing, and why is it important?
   - The significance level, often denoted by α (alpha), is the probability of rejecting the null hypothesis (H₀) when it is actually true. In simpler terms, it represents the risk of making a Type I error. It is important because:  
      - It limits the chance of making a false positive conclusion.
      - It defines the critical value or p-value threshold below which H₀ is rejected.If p-value < α, we reject the null hypothesis.
      - Provides a benchmark for researchers and decision-makers to determine whether results are statistically significant.

4. What does a P-value represent in hypothesis testing?
   - P-value in hypothesis testing is used to decide whether to reject a null hypothesis or not.
   - The p-value tells you how likely your observed data is under the assumption that the null hypothesis is true. It helps you decide whether your result is statistically significant based on a predefined significance level (α).

5. How do you interpret the P-value in hypothesis testing?
    - The p-value measures how likely your sample result is, assuming the null hypothesis is true.
    - Compare it to α to make your decision:
           - If p ≤ α, reject H₀.
           - If p > α, fail to reject H₀.

6. What are Type 1 and Type 2 errors in hypothesis testing?
   - Type 1 error is the rejection of the null hypothesis when it is actually true. You think there is an effect or difference but , there is not.
   - Type 2 error is the failure to reject the null hypothsis that is actually false. You think there is no effect or difference but there actually is.


7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
   -  One-Tailed Test: A one-tailed test is used when the alternative hypothesis predicts the direction of the effect. It checks if the sample statistic is significantly greater than or less than the hypothesized value.
       - Rejection Region: Entirely in one tail (either left or right) of the probability distribution.
       - Example:
          Null Hypothesis (H₀): μ = 50  and Alternative Hypothesis (H₁): μ > 50 (Right-tailed test)
       - This type of test is used when we are only interested in deviations in one direction.

   - Two-Tailed Test: A two-tailed test is used when the alternative hypothesis does not specify the direction of the effect. It tests whether the sample statistic is significantly different from the hypothesized value in either direction.
        - Alternative Hypothesis (H₁): Indicates a difference (e.g., not equal to).
        - Rejection Region: Split equally between both tails of the distribution.
        - Example: Null Hypothesis (H₀): μ = 50  and Alternative Hypothesis (H₁): μ ≠ 50      
        - This test is appropriate when we are interested in detecting any difference, regardless of direction.


8. What is the Z-test, and when is it used in hypothesis testing?
    - Z - test is type of hypothesis test used to determine whether there is a significant difference between sample and population means or between two sample means. It is used:
        - for large sample(n >= 30)
        - The population standard deviation is known.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
   - A Z-score (also known as the standard score) represents the number of standard deviations a data point is from the mean. It is used to compare different data points from different normal distributions and is particularly important in Z-tests, a type of hypothesis test.
   - The Z-score is calculated using the formula:
                     - Z = (X bar - mu) / (sigma/sqrt(n))

                     - where: X bar = sample mean
                              mu    = Population mean
                              sigma = Population standard deviation
                              n: Sample size

  - In hypothesis testing, the Z-score represents how many standard deviations the sample mean is away from the population mean. It helps determine whether the observed result is statistically significant. A higher absolute Z-score indicates a greater deviation from the null hypothesis. The Z-score is compared against a critical value from the Z-distribution table (based on the significance level, α) to decide whether to reject or fail to reject the null hypothesis.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
   - The T-distribution (Student’s t-distribution) is a probability distribution used in statistics when estimating population parameters when the sample size is small and/or the population standard deviation is unknown.

   - It is similar to the normal distribution but has heavier tails, meaning it accounts for more variability in small samples.
   - Use the T-distribution when:
      - The sample size is small (typically n<30)
      - The population standard deviation (σ) is unknown
      - The data is approximately normally distributed
   - Use the Normal distribution when:
       - The sample size is large (n≥30) — Central Limit Theorem applies
       - The population standard deviation is known

  - In hypothesis testing or confidence interval estimation, using the t-distribution gives more accurate results for small samples by adjusting for extra uncertainty.

     - P(A∣B)= P(B∣A)⋅P(A) / P(B)

   - P(A): Prior probability of A
   - P(B): Total probability of B
     
     - Usage in Statistics:
         - Updating beliefs: Bayes’ Theorem helps update prior knowledge after observing new data.

         - Machine learning: Forms the basis of Bayesian models and algorithms.
     - Significance:
         - Provides a mathematical framework for learning from data.
         - Shifts focus from fixed parameters (frequentist approach) to probability-based belief updates.
         - Useful when data is limited or uncertainty is involved.


  - Bayes’ Theorem is significant in statistics because it allows us to incorporate prior knowledge and update it with new evidence, making it essential in decision-making under uncertainty.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
pip install numpy
```

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = [102, 100, 98, 101, 99, 97, 103, 98, 100, 99]  # example sample
population_mean = 100   # known population mean (μ)
population_std = 2      # known population standard deviation (σ)

# Step 1: Calculate sample mean and size
sample_mean = np.mean(sample)
n = len(sample)

# Step 2: Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Step 3: Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output the result
print(f"Sample Mean: {sample_mean}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 4: Interpret the result
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Step 1: Sample and population data
np.random.seed(42)
sample = np.random.normal(loc=105, scale=15, size=40)  # Sample with mean ≈ 105
population_mean = 100
population_std = 15
n = len(sample)

# Step 2: Calculate sample mean and Z-score
sample_mean = np.mean(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Step 3: Compute two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Step 4: Decision
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)  # Two-tailed

# Step 5: Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z_score:.2f}")
print(f"P-value = {p_value:.4f}")
print(f"Z-critical = ±{z_critical:.2f}")
if abs(z_score) > z_critical:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")

# Step 6: Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution')

# Shade critical regions
plt.fill_between(x, y, where=(x < -z_critical), color='red', alpha=0.3, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x > z_critical), color='red', alpha=0.3, label='Rejection Region (Right)')

# Plot Z-score line
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_critical, color='black', linestyle='--', label=f'Z-critical = ±{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle='--')

plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    Visualizes Type I and Type II errors in hypothesis testing.

    Parameters:
    - mu0: population mean under H0
    - mu1: population mean under H1
    - sigma: population standard deviation
    - n: sample size
    - alpha: significance level (Type I error rate)
    """

    # Calculate standard error
    se = sigma / np.sqrt(n)

    # Critical value (for right-tailed test)
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    # Calculate beta (Type II error)
    z_beta = (x_crit - mu1) / se
    beta = norm.cdf(z_beta)

    # x values for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    h0_pdf = norm.pdf(x, mu0, se)
    h1_pdf = norm.pdf(x, mu1, se)

    # Plot distributions
    plt.figure(figsize=(10, 5))
    plt.plot(x, h0_pdf, label='H₀ Distribution (μ=100)', color='blue')
    plt.plot(x, h1_pdf, label='H₁ Distribution (μ=105)', color='green')

    # Shade Type I Error (alpha)
    plt.fill_between(x, h0_pdf, where=(x > x_crit), color='red', alpha=0.4, label='Type I Error (α)')

    # Shade Type II Error (beta)
    plt.fill_between(x, h1_pdf, where=(x < x_crit), color='orange', alpha=0.4, label='Type II Error (β)')

    # Critical value line
    plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')

    # Annotate
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print summary
    print(f"Critical value (sample mean cutoff): {x_crit:.2f}")
    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Call the function
visualize_type1_type2_errors()
```

8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
from scipy.stats import norm, t

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=30)  # Sample with true mean ≈ 102

# Population parameters
population_mean = 100
population_std = 15  # Known for Z-test
n = len(sample)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased)

# Step 2: Perform Z-test (population std known)
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 3: Perform T-test (population std unknown)
t_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
df = n - 1
p_value_t = 2 * (1 - t.cdf(abs(t_stat), df=df))

# Step 4: Print and compare results
print("=== Z-Test ===")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value (Z-test): {p_value_z:.4f}")

print("\n=== T-Test ===")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value (T-test): {p_value_t:.4f}")

# Step 5: Interpretation
alpha = 0.05
print("\n=== Conclusion ===")
if p_value_z < alpha:
    print("Z-test: Reject null hypothesis (significant difference).")
else:
    print("Z-test: Fail to reject null hypothesis.")

if p_value_t < alpha:
    print("T-test: Reject null hypothesis (significant difference).")
else:
    print("T-test: Fail to reject null hypothesis.")
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
    - A confidence interval (CI) gives a range of values within which the true population mean is likely to fall, with a specified level of confidence (e.g., 95%). It reflects the uncertainty associated with using a sample to estimate a population parameter.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.95)

    Returns:
    - (lower_bound, upper_bound): tuple of confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Sample standard deviation
    se = std_dev / np.sqrt(n)       # Standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_crit * se
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return (lower_bound, upper_bound), mean

# Example usage:
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=30)
(ci_range, mean_val) = confidence_interval(sample_data)

print(f"Sample Mean = {mean_val:.2f}")
print(f"95% Confidence Interval = ({ci_range[0]:.2f}, {ci_range[1]:.2f})")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.95)

    Returns:
    - margin_of_error (float)
    """
    n = len(data)
    std_dev = np.std(data, ddof=1)         # Sample standard deviation
    se = std_dev / np.sqrt(n)              # Standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)  # t-critical value

    margin_of_error = t_crit * se
    return margin_of_error

# Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=100, scale=12, size=40)  # Simulated sample
confidence_level = 0.95

moe = calculate_margin_of_error(sample_data, confidence=confidence_level)
print(f"Sample Size: {len(sample_data)}")
print(f"Confidence Level: {int(confidence_level * 100)}%")
print(f"Margin of Error: ±{moe:.2f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayesian_inference(prior, likelihood, false_positive_rate):
    """
    Apply Bayes' Theorem to compute the posterior probability.

    Parameters:
    - prior: P(H), prior probability of hypothesis (e.g., having disease)
    - likelihood: P(E|H), probability of evidence given hypothesis (e.g., positive test if diseased)
    - false_positive_rate: P(E|¬H), probability of evidence given hypothesis is false (false alarm)

    Returns:
    - posterior: P(H|E), updated probability after seeing evidence
    """
    not_prior = 1 - prior
    evidence = (likelihood * prior) + (false_positive_rate * not_prior)
    posterior = (likelihood * prior) / evidence
    return posterior

# Example values
prior = 0.01                  # P(Disease)
likelihood = 0.95             # P(Positive | Disease)
false_positive = 0.05         # P(Positive | No Disease)

posterior = bayesian_inference(prior, likelihood, false_positive)

print(f"Posterior Probability (P(Disease | Positive Test)): {posterior:.4f}")
```

 Explanation of the Process:
    - Define Prior: What we initially believe (e.g., 1% chance of disease).
    - Define Likelihood: How likely the evidence is if the hypothesis is true (e.g., test accuracy).
    - Define False Positive Rate: Likelihood of seeing the evidence even when the hypothesis is false.
    - Use Bayes' Theorem: Update the prior using the new evidence to compute the posterior.

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

# Step 1: Simulate two independent samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # std = 10
sample2 = np.random.normal(loc=55, scale=15, size=30)  # std = 15

# Step 2: Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Step 3: Compute F-statistic (larger variance / smaller variance)
if var1 > var2:
    F = var1 / var2
    df1, df2 = len(sample1)-1, len(sample2)-1
else:
    F = var2 / var1
    df1, df2 = len(sample2)-1, len(sample1)-1

# Step 4: Calculate p-value (two-tailed test)
p_value = 2 * min(f.cdf(F, df1, df2), 1 - f.cdf(F, df1, df2))

# Step 5: Print results
print("Sample 1 Variance:", round(var1, 2))
print("Sample 2 Variance:", round(var2, 2))
print("F-statistic:", round(F, 2))
print("Degrees of Freedom:", df1, "and", df2)
print("P-value:", round(p_value, 4))

# Step 6: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis — variances are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis — no significant difference in variances.")
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, levene
import statsmodels.api as sm
from statsmodels.formula.api import ols

def check_anova_assumptions(data, value_col, group_col):
    """
    Check ANOVA assumptions: normality, independence, and equal variance.

    Parameters:
    - data: pandas DataFrame containing the data
    - value_col: name of the column with numerical values
    - group_col: name of the column with group labels
    """

    # Grouped data
    groups = data[group_col].unique()

    print("=== 1. Normality Test (Shapiro-Wilk per group) ===")
    for group in groups:
        sample = data[data[group_col] == group][value_col]
        stat, p = shapiro(sample)
        print(f"{group}: W={stat:.4f}, p={p:.4f} → {'Pass' if p > 0.05 else 'Fail'}")

    print("\n=== 2. Equal Variance Test (Levene’s Test) ===")
    group_values = [data[data[group_col] == g][value_col] for g in groups]
    stat, p = levene(*group_values)
    print(f"Levene’s test: W={stat:.4f}, p={p:.4f} → {'Pass' if p > 0.05 else 'Fail'}")

    print("\n=== 3. Independence Check (Residual Plot) ===")
    model = ols(f"{value_col} ~ C({group_col})", data=data).fit()
    residuals = model.resid
    fitted = model.fittedvalues

    plt.figure(figsize=(8, 5))
    sns.residplot(x=fitted, y=residuals, lowess=True, color="blue", line_kws={'color': 'red'})
    plt.axhline(0, color='gray', linestyle='--')
    plt.title("Residuals vs Fitted Values")
    plt.xlabel("Fitted Values")
    plt.ylabel("Residuals")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    print("→ Look for no clear pattern (random scatter) to confirm independence.")


# Simulate example data
np.random.seed(10)
df = pd.DataFrame({
    'Score': np.concatenate([
        np.random.normal(60, 5, 30),
        np.random.normal(65, 5, 30),
        np.random.normal(70, 5, 30)
    ]),
    'Group': ['A']*30 + ['B']*30 + ['C']*30
})

# Run assumption checks
check_anova_assumptions(df, value_col='Score', group_col='Group')
```

24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=100, scale=12, size=25)  # std=12, so variance=144
n = len(sample)

# Step 2: Known or hypothesized population variance
pop_variance = 100  # Assume σ₀² = 100

# Step 3: Calculate sample variance
sample_variance = np.var(sample, ddof=1)

# Step 4: Compute Chi-square test statistic
chi_square_stat = (n - 1) * sample_variance / pop_variance

# Step 5: Critical values and p-value
alpha = 0.05
df = n - 1
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)
p_value = 2 * min(chi2.cdf(chi_square_stat, df), 1 - chi2.cdf(chi_square_stat, df))

# Step 6: Print Results
print("=== Chi-square Test for Variance ===")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Hypothesized Variance: {pop_variance}")
print(f"Chi-square Statistic: {chi_square_stat:.2f}")
print(f"Degrees of Freedom: {df}")
print(f"Chi-square Critical Range: [{chi2_lower:.2f}, {chi2_upper:.2f}]")
print(f"P-value: {p_value:.4f}")

# Step 7: Conclusion
if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
    print("Conclusion: Reject H₀ — the variance is significantly different from the population variance.")
else:
    print("Conclusion: Fail to reject H₀ — no significant difference in variance.")

# Step 8: Optional Visualization
x = np.linspace(chi2.ppf(0.001, df), chi2.ppf(0.999, df), 500)
y = chi2.pdf(x, df)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Chi-square Distribution')
plt.axvline(chi_square_stat, color='red', linestyle='--', label='Test Statistic')
plt.axvline(chi2_lower, color='green', linestyle=':', label='Critical Lower')
plt.axvline(chi2_upper, color='green', linestyle=':', label='Critical Upper')
plt.title('Chi-square Test for Population Variance')
plt.xlabel('Chi-square Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Step 1: Simulate observed data (frequency of purchases in 4 categories)
np.random.seed(123)
observed = np.array([55, 65, 50, 30])  # e.g., number of customers preferring 4 products

# Step 2: Define expected frequencies (uniform distribution assumed here)
expected = np.full_like(observed, fill_value=np.mean(observed))

# Step 3: Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Print results
print("Chi-square Goodness-of-Fit Test ")
print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"Chi-square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Conclusion
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H₀ — the observed distribution does NOT fit the expected.")
else:
    print("Conclusion: Fail to reject H₀ — the observed distribution fits the expected.")

# Step 6: Visualization
categories = ['Product A', 'Product B', 'Product C', 'Product D']
x = np.arange(len(categories))

plt.figure(figsize=(8, 5))
plt.bar(x - 0.15, observed, width=0.3, label='Observed', color='skyblue')
plt.bar(x + 0.15, expected, width=0.3, label='Expected', color='lightgreen')
plt.xticks(x, categories)
plt.title("Observed vs Expected Frequencies (Chi-square Test)")
plt.xlabel("Category")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
