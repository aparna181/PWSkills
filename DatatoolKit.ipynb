2. How does broadcasting work in NumPy?
   - Broadcasting in NumPy is a technique that allows arrays of different shapes to be used together in arithmetic operations. Instead of resizing arrays manually, NumPy automatically stretches the smaller array across the larger one so their shapes become compatible. This is done without actually copying the data, which makes operations faster and memory-efficient. For broadcasting to work, the dimensions of the arrays must either be equal or one of them must be 1. If this condition is met, NumPy performs the operation as if both arrays had the same shape. Broadcasting is especially useful for combining scalars or lower-dimensional arrays with higher-dimensional ones in a simple and readable way.

```python
!pip install numpy
```

```python
import numpy as np

a = np.array([1, 2, 3])       # Shape: (3,)
b = np.array([[10], [20]])    # Shape: (2, 1)

result = a + b                # Shape: (2, 3)
print(result)

```

Here, a is broadcast to match the shape of b, resulting in a 2x3 matrix.

8. What does the term “vectorized operation” mean in NumPy?
   - Vectorized operation in NumPy means performing element-wise calculations on whole arrays at once, without using explicit loops. This makes the code faster and cleaner because NumPy runs operations in optimized, low-level code.

17.  Explain the difference between apply() and map() in Pandas.
     - apply() is used to apply a function to each element, row, or column of a DataFrame or Series. It works on both Series and DataFrames and can handle more complex functions.

       map() is mainly used to substitute or map values in a Series based on a dictionary, function, or Series. It works only on Series and is ideal for simple value replacements.



```python
import pandas as pd

# Sample Series
s = pd.Series([1, 2, 3, 4])

# Using map() to replace values
mapped = s.map({1: 'one', 2: 'two', 3: 'three'})
print("Using map():\n", mapped)

# Using apply() to square each value
applied = s.apply(lambda x: x**2)
print("\nUsing apply():\n", applied)
```

19. How does Pandas simplify time series analysis?
    - Pandas makes time series analysis easier by providing tools to work with dates and times, such as:

     DateTime indexing to access and slice data by date or time.

     Resampling to change the frequency of data (e.g., daily to monthly).

     Time-shifting to move data forward or backward in time.

     Handling missing dates automatically.

     Built-in functions for rolling windows, moving averages, and time-based grouping

```python
import pandas as pd

# Create a time series with daily data
dates = pd.date_range('2023-01-01', periods=6, freq='D')
data = pd.Series([10, 12, 15, 14, 13, 16], index=dates)

# Resample daily data to monthly data by taking the mean
monthly_data = data.resample('M').mean()

print("Original daily data:")
print(data)

print("\nResampled monthly data:")
print(monthly_data)
```

```python
#example time shifting
import pandas as pd

# Create a time series
dates = pd.date_range('2023-01-01', periods=5, freq='D')
data = pd.Series([5, 10, 15, 20, 25], index=dates)

# Shift data forward by 1 day
shifted = data.shift(1)

print("Original Data:\n", data)
print("\nShifted Data (1 day forward):\n", shifted)
```

```python
#Rolling window
# Calculate 3-day moving average
rolling_avg = data.rolling(window=3).mean()

print("\n3-Day Moving Average:\n", rolling_avg)
```

3. Create a scatter plot using Matplotlib.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target as a column
df['species'] = iris.target

# Create scatter plot: Sepal Length vs Sepal Width
plt.figure(figsize=(6, 4))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['species'], cmap='viridis')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.colorbar(label='Species')
plt.grid(True)
plt.show()
```

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris dataset from Seaborn
df = sns.load_dataset("iris")

# Calculate correlation only on numeric columns (exclude 'species')
correlation_matrix = df.select_dtypes(include=['float64', 'int']).corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Iris Dataset")
plt.show()
```

5. Generate a bar plot using Plotly

```python
import plotly.express as px
import seaborn as sns
import pandas as pd

# Load Iris dataset from seaborn
df = sns.load_dataset("iris")

# Calculate average sepal length for each species
avg_sepal_length = df.groupby('species')['sepal_length'].mean().reset_index()

# Create a bar plot with Plotly Express
fig = px.bar(avg_sepal_length, x='species', y='sepal_length',
             title='Average Sepal Length by Iris Species',
             labels={'sepal_length': 'Average Sepal Length (cm)', 'species': 'Species'})

fig.show()
```

11. Perform matrix multiplication using NumPy.

```python
import numpy as np

# Define two 2D arrays (matrices)
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Matrix multiplication using np.dot() or @ operator
result = np.dot(A, B)
# Or equivalently: result = A @ B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix multiplication result:\n", result)
```
