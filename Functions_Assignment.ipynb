1. What is the difference between a function and a method in Python?
 - A function is a block of code that is defined using the def keyword and can be called independently by its name. It is not associated with any object and can be used globally in the program.

```python
def add_numbers(a, b):
    return a + b

# Calling the function
result = add_numbers(10, 20)
print("Sum is:", result)
```

   -  A method is also a function, but it is associated with an object. Methods are defined inside a class and are called using the object of that class. The first parameter of a method is always self, which refers to the instance of the class.

```python
class Calculator:
    def add(self, a, b):
        return a + b

# Creating an object of the class
calc = Calculator()

# Calling the method
result = calc.add(10, 20)
print("Sum is:", result)
```

5. What are iterators in Python and how do they differ from iterables?
- Iterable: An object capable of returning its elements one at a time. Examples include lists, tuples, strings, and sets.

 Iterator: An object representing a stream of data; it returns the next value when you call next() on it. It implements both __iter__() and __next__() methods.

 Difference: Iterable is a container that can be looped over. Iterator is the object used to perform the actual iteration. Using iter() on an iterable gives an iterator. Iterators maintain their state and can be exhausted; iterables can be reused to create new iterators.

```python
# Iterable
my_list = [1, 2, 3]

# Converting iterable to iterator
my_iter = iter(my_list)

# Accessing elements using next()
print(next(my_iter))
print(next(my_iter))
```

6. Explain the concept of generators in Python and how they are defined.
   - A generator in Python is a special type of iterable that allows you to iterate over data without storing it all in memory at once. It generates items one at a time using the yield keyword, which makes it more memory-efficient for large datasets.

```python
# Defining a generator function
def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # Yielding the next value
        count += 1

# Using the generator
counter = count_up_to(5)
for num in counter:
    print(num)
```

7. What are the advantages of using generators over regular functions?
   - Memory Efficiency: Generators generate values one at a time and do not store all values in memory, making them ideal for handling large datasets.
   
    Lazy Evaluation: Values are produced only when requested, which can improve performance when dealing with large or infinite data streams.

    State Retention: Generators remember their state between yields, allowing them to produce values incrementally without needing to recompute.  

example of using Generator:

```python
def even_numbers(max):
    n = 2
    while n <= max:
        yield n
        n += 2

gen = even_numbers(10)
for num in gen:
    print(num)
```

Regular Function:

```python
def even_numbers_list(max):
    numbers = []
    n = 2
    while n <= max:
        numbers.append(n)
        n += 2
    return numbers

nums = even_numbers_list(10)
print(nums)
```

9. Explain the purpose and usage of the 'map()' function in Python.
-  The map() function in Python is used to apply a given function to all items in an iterable (like a list, tuple, etc.) and return a new map object (which is an iterator) with the results.

   The main purpose of map() is to simplify code where you want to perform a transformation on each element of a collection.

   syntax:  map(function, iterable)


```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
print(list(squared))
```

```python
#Using map() with multiple iterables
a = [1, 2, 3]
b = [4, 5, 6]
result = map(lambda x, y: x + y, a, b)
print(list(result))
```

10. What is the difference between 'map()', 'reduce()', and 'filter()' functions in Python?
  

 -  a. map():  Applies a function to each element of an iterable and returns a new iterable (map object) with the results.

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
print(list(squared))
```

 - b.filter():  Filters elements of an iterable based on a condition (i.e., a function that returns True or False).

```python
numbers = [1, 2, 3, 4, 5]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))
```

-  c. reduce(): Applies a function cumulatively to the elements of an iterable, reducing it to a single value.

```python
from functools import reduce
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

 Difference: The map() function is used to transform each item in an iterable, producing an output of the same length as the input, using any transformation function. The filter() function selects items from an iterable based on a condition, resulting in an output that is possibly shorter, and it requires a function that returns a boolean value. On the other hand, the reduce() function collapses all items in an iterable into a single value by applying a cumulative operation, such as summing or multiplying elements.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

nums = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(nums)
print("Sum of even numbers:", result)
```

4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True


number = int(input("Enter a number between 1 and 200: "))

if 1 <= number <= 200:
    if is_prime(number):
        print(number, "is a prime number.")
    else:
        print(number, "is not a prime number.")
else:
    print("Please enter a number within the range 1 to 200.")
```

terms.

```python
def fib(n):
    a, b = 0, 1
    count = 0
    while count < n:
        yield a
        a, b = b, a + b
        count += 1

f = iter(fib(10))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
```

6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers(exponent):
    for i in range(exponent + 1):
        yield 2 ** i


exponent = int(input("Enter the exponent: "))
for power in powers(exponent):
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
  """Generator function that reads a file line by line and yields each line
   as a string"""
  try:
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
  except FileNotFoundError:
    raise FileNotFoundError(f"The file{file_path} was not found.")


file_path = '/content/pytthon..txt'
for line in read_file_line_by_line(file_path):
    print(line)
```

10. Create a Python program that uses 'filter()' to remove all the vowels from a given string.

```python
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    result = ''.join(filter(lambda char: char not in vowels, input_string))
    return result

text = "Hello World, Python is amazing!"
filtered_text = remove_vowels(text)

print(filtered_text)
```

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
order is smaller than 100,00 .

```python
# List of orders: (Order Number, Book Title and Author, Quantity, Price per Item)
orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99)
]
result = []
for order in orders:
    order_num = order[0]
    quantity = order[2]
    price_per_item = order[3]

    total = quantity * price_per_item

    # Add 10  if the total is less than 100 
    if total < 100:
        total += 10

    result.append((order_num, round(total, 2)))
print(result)
```
