1. What is statistics, and why is it important?

2. What are the two main types of statistics?

3. What are descriptive statistics?
   - Descriptive statistics refers to the statistical methods that are used to describe and summarize data. It includes the use of numbers, tables, charts, and graphs to present the essential features of data. It is primarily concerned with what the data shows.

    It involves 3 techniques:
   
    (a)Measure of central tendency(mean,median,mode)

    (b)Measure of Symmetricity(skewness,kurtosis)
    
    (c)Measure if Dispersion(variance,range,standard deviation)


4. What is inferential statistics?

    It tries to draw conclusions and test hypotheses using probability theory.

5. What is sampling in statistics?

This is crucial when working with large datasets as it allows data scientist to make inferences about the population without analysiing the entire dataset, which may be impractical or resource-intensive.

8.  Define and give examples of qualitative and quantitative data.

9. What are the different types of data in statistics?
   - There is two types of data. First one is Qualitative data known as categorical data and second one is quantitative data known as numerical data .
    
    Further numerical data divide into two parts: Discreate data and continuous data .
    
       - Discreate data : Consists of whole numbers; values are countable and finite. Examples: Number of students in a class, Number of cars, Shoe sizes.

       - Continous data : Can take any value within a range, including decimals; data is measurable. Examples: Height, Weight, Temperature, Time

10. Explain nominal, ordinal, interval, and ratio levels of measurement.
   - **Nominal Level of measurement:** Nominal data refers to categories or labels without any order or ranking. Ex-Gender(Male, Female), Blood Type( A, B, AB, O), Marital Status( Single, Married, Divorced),Hair Color(Black, Brown, Blonde)

        - Only classification is possible

        - No logical order

        - Cannot perform arithmetic operations

        - To show using pie chart, bar chart

     **Ordinal Level of measurement:** Ordinal data involves categories with a meaningful order or ranking, but the differences between the values are not measurable. Ex- Educational Level(High School < Bachelor < Master < PhD),Customer Satisfaction(Poor, Average, Good, Excellent),Class Rank(1st, 2nd, 3rd)

        - Order matters

        - No consistent difference between values

        - Cannot do accurate arithmetic operations
        - To show using pie chart, bar chart

      **Interval Level:** Interval data includes numerical values where the difference between values is meaningful, but there is no true zero. Ex- Temperature in Celsius or Fahrenheit, Calendar years (e.g., 2000, 2025)
         - Equal intervals between values
         - No absolute zero (zero doesn’t mean absence)
         - Can add and subtract values
         - using scatter plots and linechart , histogram

      **Ratio Level:**  Ratio data is the highest level of measurement. It includes all properties of interval data, plus a true zero, allowing for meaningful ratios. Ex- Height, Weight, Age, Distance, Salary, Time duration

      - True zero exists (zero means absence)
      - Can add, subtract, multiply, and divide
      - All mathematical operations are valid

18. What is skewness in a dataset?
    - Skewness in a dataset refers to the asymmetry or lack of symmetry in the distribution of data values around the mean.

        - A positive skew (right-skewed) means the tail is longer on the right side.
        - A negative skew (left-skewed) means the tail is longer on the left side.
        - Zero skewness means the data is perfectly symmetrical (normal distribution).

19. What does it mean if a dataset is positively or negatively skewed?

       - Mean > Median > Mode

   - A negatively skewed dataset means the tail on the left side is longer, and most data values are concentrated on the right (higher values).
      - Mean < Median < Mode

20. Define and explain kurtosis.
    - Kurtosis is a statistical measure that describes the shape of a distribution's tails in relation to its overall shape—specifically, the "peakedness" or flatness of the distribution compared to a normal distribution.

  - Types of Kurtosis:
        - Mesokurtic (Normal Kurtosis): Distribution has a moderate peak. Example: Normal distribution
        
         Kurtosis ≈ 3

        - Leptokurtic (High Kurtosis):Distribution has a sharp peak and heavy tails (more outliers).

           Kurtosis > 3

      - Platykurtic (Low Kurtosis):Distribution has a flat peak and light tails (fewer outliers).

           Kurtosis < 3

22. What does correlation measure in statistics?
    - Correlation in statistics measures the strength and direction of a linear relationship between two variables.

    - It is expressed using a value called the correlation coefficient (r), which ranges from -1 to +1:

     +1 → perfect positive correlation

     0 → no correlation

      –1 → perfect negative correlation

    - In simple terms, it shows how one variable changes in relation to another.

23. What is the difference between covariance and correlation?
    - Covariance measures the direction of the relationship between two variables, while correlation measures both the strength and direction.

    - The range of covariance is : from –∞ to +∞ and The range of correlation is : from -1 to +1.
    - Covariance is not standardized. Its value depends on the units of the variables involved. For example, if one variable is in kilograms and another in centimeters, the covariance will be in kg·cm — making it hard to interpret or compare. Correlation is the standardized form of covariance. It removes units by dividing the covariance by the product of the standard deviations of the two variables. This results in a value between –1 and +1, making it easier to interpret and compare across datasets.

13. Calculate the mean, median, and mode of grouped data.

```python
# Load Iris dataset
iris = sns.load_dataset('iris')
```

```python
# Calculate mean, median, and mode of original sepal_length (not grouped)
mean_val = iris['sepal_length'].mean()
median_val = iris['sepal_length'].median()
mode_val = iris['sepal_length'].mode()

print(f"\nMean of Sepal Length: {mean_val:.2f}")
print(f"Median of Sepal Length: {median_val:.2f}")
print(f"Mode of Sepal Length: {list(mode_val)}")
```

```python
# Group sepal_length into intervals (bins)
bins = [4, 5, 6, 7, 8]
labels = ['4-5', '5-6', '6-7', '7-8']
iris['sepal_length_group'] = pd.cut(iris['sepal_length'], bins=bins, labels=labels, right=False)

# Frequency table
freq_table = iris['sepal_length_group'].value_counts().sort_index()
print("Frequency Table:\n", freq_table)

# Midpoint of each group (for mean calculation)
midpoints = [4.5, 5.5, 6.5, 7.5]

# Calculate mean of grouped data
mean_grouped = np.sum(freq_table.values * midpoints) / freq_table.sum()
print(f"\nMean (Grouped Data): {mean_grouped:.2f}")

# To approximate median:
# 1. Calculate cumulative frequency
cum_freq = freq_table.cumsum()
N = freq_table.sum()
median_class_index = cum_freq.searchsorted(N/2)
median_class = labels[median_class_index]
print(f"Median Group: {median_class}")

# For mode:
mode_class = freq_table.idxmax()
print(f"Mode Group: {mode_class}")
```

14.  Simulate data using Python and calculate its central tendency and dispersion.

23. Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Load the tips dataset
tips = sns.load_dataset('tips')

data = tips['total_bill']

# Calculate skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data)  # excess kurtosis (normal dist = 0)

print(f"Skewness: {data_skewness:.3f}")
print(f"Kurtosis: {data_kurtosis:.3f}")

# Plot histogram with KDE to visualize distribution shape
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30, color='skyblue')

# Add vertical lines for mean and median for reference
plt.axvline(data.mean(), color='red', linestyle='--', label='Mean')
plt.axvline(data.median(), color='green', linestyle='-', label='Median')

plt.title(f"Histogram & KDE of Total Bill\nSkewness: {data_skewness:.3f}, Kurtosis: {data_kurtosis:.3f}")
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
