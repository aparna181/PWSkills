1. What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning?
     -  Deep Learning is a subset of machine learning (ML) that uses artificial neural networks with multiple layers (hence “deep”) to automatically learn patterns and representations from large amounts of data. It excels at handling unstructured data like images, audio, text, and video.
     - 1940s–1960s: Early neural networks were inspired by biological neurons (e.g., McCulloch-Pitts model).
     - 1980s: Introduction of backpropagation for training multi-layer networks, allowing deeper networks.
     - 1990s–2000s: Neural networks faced challenges like vanishing gradients and limited computational power. ML methods like SVMs and decision trees dominated.
     - 2010s–present: Advances in GPUs, large datasets, and architectures like CNNs (Convolutional Neural Networks) and RNNs (Recurrent Neural Networks) led to breakthroughs in image, speech, and natural language processing. DL became practical and outperforming traditional ML in many domains.
     
     - Deep Learning (DL) differs from traditional Machine Learning (ML) in several key aspects. In traditional ML, feature extraction is manual and requires domain expertise, whereas in deep learning, features are automatically learned from raw data. Traditional ML performs well on small to medium-sized datasets, while deep learning typically requires large datasets to achieve optimal performance. In terms of applicability, traditional ML is suitable for structured or tabular data, whereas deep learning excels at handling unstructured data such as images, audio, and text. Traditional ML models, like decision trees and SVMs, are relatively simple, while deep learning relies on complex multi-layer neural networks. Finally, traditional ML is less computationally intensive, whereas deep learning demands high computational resources, often leveraging GPUs or TPUs for efficient training.

2. Explain the basic architecture and functioning of a Perceptron. What are its
limitations?
    - A Perceptron is the simplest type of artificial neural network, introduced by Frank Rosenblatt (1958). It is a binary classifier that decides whether an input belongs to one class or another.
    - A perceptron consists of:
       - Input : The features of the data.
       - Weights : Each input has an associated weight representing its importance.
       - Bias (b): A threshold value that helps the model shift the decision boundary.
       - Summation function: Computes the weighted sum of inputs
       - Activation function: Applies a step function to produce output y .      
         
         y=if z≥0 , then 1 and if z<0, then 0

     - Limitations of Perceptron:
          - Cannot solve non-linear problems: For example, it fails on XOR problem because it can only learn linearly separable functions.
          - Single layer only: Classic perceptrons do not have hidden layers, limiting their ability to learn complex patterns.
          - Binary output: Standard perceptrons only produce 0 or 1, making them unsuitable for multi-class or continuous output tasks.
          - Sensitive to scaling: Inputs often need normalization for better convergence.

3. Describe the purpose of activation function in neural networks. Compare
Sigmoid, ReLU, and Tanh functions.
   - Introduces non-linearity into neural networks.
   - Allows the network to learn complex patterns beyond linear relationships.
   - Determines whether a neuron should be activated based on its input.
   - Helps in controlling the output range and stabilizing learning.

   - Sigmoid:
      - Output range: 0 to 1
      - Characteristics: Smooth S-shaped curve
      - Advantages: Output interpretable as probability
      - Limitations: Vanishing gradient for large positive/negative inputs, Slow convergence

   - Tanh:
       - Output range: -1 to 1
       - Characteristics: S-shaped curve, centered at 0
       - Advantages: Zero-centered output, Faster convergence than sigmoid
       - Limitations: Still suffers from vanishing gradient for large inputs   
   - ReLU (Rectified Linear Unit):
       - Output range: 0 to ∞
       - Characteristics: Linear for positive values, 0 for negatives
       - Advantages: Fast computation, Alleviates vanishing gradient problem, Produces sparse activations
       - Limitations: “Dying ReLU” problem (neurons stuck at 0), Not zero-centered

4. What is the difference between Loss function and Cost function in neural
networks? Provide examples.
- Loss Function:
     - Definition: Measures the error for a single training example.
     - Purpose: Tells how far the model’s prediction is from the actual target for one data point.
- Cost Function:
     - Definition: The average of the loss function over the entire training dataset (or a batch).
     - Purpose: Gives an overall measure of model performance during training.   


5. What is the role of optimizers in neural networks? Compare Gradient
Descent, Adam, and RMSprop.
- Optimizers are algorithms used to update the weights and biases of a neural network to minimize the cost function.

- They guide the model during training by deciding how much and in what direction the parameters should change.

- GD : Gradient Descent is the simplest optimizer that updates weights by moving in the direction of the negative gradient of the cost function. While it is easy to implement and mathematically stable, it can be slow to converge and is highly sensitive to the choice of learning rate.
- RMSprop : It  improves upon this by adapting the learning rate for each parameter individually, using a moving average of squared gradients. This makes it particularly effective for problems with non-stationary objectives, such as recurrent neural networks, though it can be sensitive to hyperparameter settings.
- Adam : Adam combines the advantages of both momentum (which smooths the gradient updates) and RMSprop (which adapts learning rates), making it one of the most widely used optimizers today. It converges faster, requires little tuning, and works well across different types of problems, although it may sometimes lead to overfitting or poorer generalization compared to SGD.

6. Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.


```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


# Part 1: Perceptron from scratch using NumPy


# Input (AND gate truth table)
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

# Output
y = np.array([0, 0, 0, 1])

# Initialize weights and bias
np.random.seed(42)
weights = np.random.rand(2)
bias = np.random.rand(1)

# Learning rate
lr = 0.1

# Activation function (step function)
def step_function(z):
    return np.where(z >= 0, 1, 0)

# Training the perceptron
epochs = 20
errors = []

for epoch in range(epochs):
    total_error = 0
    for i in range(len(X)):
        # Weighted sum
        z = np.dot(X[i], weights) + bias
        y_pred = step_function(z)

        # Update rule
        error = y[i] - y_pred
        weights = weights + lr * error * X[i]
        bias = bias + lr * error

        total_error += abs(error)
    errors.append(total_error)
    print(f"Epoch {epoch+1}/{epochs}, Total Error: {total_error}")

print("\nFinal Weights:", weights)
print("Final Bias:", bias)

# Predictions
print("\nPredictions after training:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    print(f"Input: {X[i]}, Output: {step_function(z)}")

# Plot error over epochs
plt.plot(range(1, epochs+1), errors, marker='o')
plt.title("Training Error over Epochs (NumPy Perceptron)")
plt.xlabel("Epochs")
plt.ylabel("Total Error")
plt.show()

# Part 2: Implementation using TensorFlow/Keras

model = Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))  # sigmoid for binary classification

model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X, y, epochs=100, verbose=0)

# Evaluate
loss, acc = model.evaluate(X, y, verbose=0)
print(f"Keras Model Accuracy: {acc*100:.2f}%")

# Predictions
y_pred_keras = (model.predict(X) > 0.5).astype("int32")
for i in range(len(X)):
    print(f"Input: {X[i]}, Output: {y_pred_keras[i][0]}")
```

7. Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Input range
x = np.linspace(-10, 10, 200)

# Apply activation functions
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plotting
plt.figure(figsize=(12, 4))

# Sigmoid
plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, 'b')
plt.title("Sigmoid Function")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("f(x)")

# ReLU
plt.subplot(1, 3, 2)
plt.plot(x, y_relu, 'r')
plt.title("ReLU Function")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("f(x)")

# Tanh
plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, 'g')
plt.title("Tanh Function")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("f(x)")

plt.tight_layout()
plt.show()
```

8.  Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 1. Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values (0–255 → 0–1)
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)


# 2. Build the Neural Network

model = Sequential([
    Flatten(input_shape=(28, 28)),          # Flatten 28x28 images into 784 vector
    Dense(128, activation='relu'),          # Hidden layer
    Dense(64, activation='relu'),           # Another hidden layer
    Dense(10, activation='softmax')         # Output layer (10 classes)
])


# 3. Compile the model

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 4. Train the model

history = model.fit(X_train, y_train,
                    epochs=5,
                    batch_size=128,
                    verbose=1)

# 5. Print Training Accuracy

train_acc = history.history['accuracy'][-1]
print(f"\nFinal Training Accuracy: {train_acc*100:.2f}%")
```

9. Visualize the loss and accuracy curves for a neural network model trained
on the Fashion MNIST dataset. Interpret the training behavior.


```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 1. Load and preprocess data
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize pixel values
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)


# 2. Build the Neural Network

model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])
# 3. Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 4. Train the model
history = model.fit(X_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_data=(X_test, y_test),
                    verbose=1)

# 5. Plot Loss and Accuracy
plt.figure(figsize=(12, 5))

# Plot Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curve")
plt.legend()
plt.grid(True)

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Accuracy Curve")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

10.  You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured
features like transaction amount, merchant ID, and customer location. The goal is to
classify each transaction as fraudulent or legitimate.
Explain your real-time data science workflow:
● How would you design a deep learning model (perceptron or multilayer NN)?
● Which activation function and loss function would you use, and why?
● How would you train and evaluate the model, considering class imbalance?
● Which optimizer would be suitable, and how would you prevent overfitting?

1. Designing the Deep Learning Model

To detect fraudulent transactions, a multilayer perceptron (MLP) is more suitable than a single perceptron because the dataset contains many structured features (transaction amount, merchant ID, location, etc.) with complex non-linear relationships.

Input: Numerical features are normalized, categorical features (like merchant ID) can be encoded using embeddings or one-hot encoding.

Architecture: Input layer → 2–3 hidden dense layers (e.g., 128 → 64 → 32 neurons) with ReLU activation → Output layer with 1 neuron and sigmoid activation for binary classification (fraud or legitimate).

2. Activation Function and Loss Function

Hidden layers: ReLU is used because it reduces the vanishing gradient problem and speeds up convergence.

Output layer: Sigmoid is chosen to output probabilities between 0 and 1.

Loss function: Binary Cross-Entropy is the standard choice for binary classification. However, due to severe class imbalance, Weighted Binary Cross-Entropy or Focal Loss is preferred to give more importance to minority fraud cases.

3. Training and Evaluation with Class Imbalance

Data splitting: Use time-based train–validation splits to avoid information leakage.

Class imbalance handling: Assign class weights in the loss function so that fraudulent cases have higher penalty. Alternatively, apply oversampling (SMOTE) for the minority class or undersampling for the majority class. Focal loss can be used to focus learning on hard-to-classify fraud cases.

Evaluation metrics: Accuracy is misleading. Instead, metrics like Precision, Recall, F1-score, ROC AUC, and Precision-Recall AUC are used. For fraud detection, Recall (catching more fraud) and Precision (reducing false alarms) are most important.

4. Optimizer and Overfitting Prevention

Optimizer: Adam is a good choice because it adapts learning rates automatically and works well on sparse, noisy data. RMSprop can also be considered for stability, but Adam generally converges faster.

Preventing overfitting: Use Dropout layers and L2 regularization. Apply Early Stopping based on validation loss or AUC. Monitor for data leakage and remove features that directly reveal fraud labels.
