#Theoretical Questions:

1. What is unsupervised learning in the context of machine learning?
   - Unsupervised learning in machine learning is a type of learning where the model is trained on unlabeled data, meaning the dataset does not contain predefined outputs or target variables. Instead of learning from input–output pairs (like in supervised learning), the algorithm tries to find hidden patterns, groupings, or structures in the data on its own.

2. How does K-Means clustering algorithm work?
   - K-Means is one of the most popular unsupervised learning algorithms for clustering. It groups data points into K clusters based on similarity.
   - Step-by-step working of K-Means:
   1. Choose number of clusters (K):Decide how many groups you want to divide the data into.
   2. Initialize centroids:Randomly pick K data points (or use methods like k-means++) as the initial centroids (the center of each cluster).
   3. Assign data points to nearest centroid:For each data point, calculate the distance (commonly Euclidean distance) to all centroids and assign it to the nearest centroid’s cluster.
   4. Update centroids: After all points are assigned, recalculate the centroid of each cluster as the mean of all points in that cluster.
   5. Repeat steps 3 & 4 until convergence:Assign points again to nearest centroids.Recalculate centroids.
Stop when centroids no longer change (or changes are minimal) → convergence

3. Explain the concept of a dendrogram in hierarchical clustering.
   - A dendrogram is a tree-like diagram used in hierarchical clustering to represent the arrangement of clusters formed at different stages of the algorithm.
     - In hierarchical clustering, each data point starts as its own cluster.
     - At every step, the two closest clusters are merged based on a distance metric (e.g., Euclidean distance).
     - This merging process continues until all points are grouped into a single cluster.

   - In a dendrogram:
       - The x-axis represents the data points or clusters.
       - The y-axis represents the distance (or dissimilarity) at which clusters are combined.
       - The vertical lines show when clusters are merged, and the height indicates the distance between them.

A dendrogram is especially useful for deciding the number of clusters by drawing a horizontal cut across the tree; the number of vertical lines intersected gives the number of clusters.

4. What is the main difference between K-Means and Hierarchical Clustering?
   - The main difference between K-Means and Hierarchical Clustering is in how clusters are formed:
       - K-Means Clustering is a partitioning method where you predefine the number of clusters (K), and the algorithm assigns data points to clusters by minimizing the distance to the cluster centroids.
       - Hierarchical Clustering is a tree-based method that builds a hierarchy of clusters without needing to predefine K. It can be agglomerative (bottom-up) or divisive (top-down), and the result is represented using a dendrogram.

5. What are the advantages of DBSCAN over K-Means?

    - **Advantages of DBSCAN over K-Means:**
    1. No need to specify number of clusters (K): K-Means requires you to predefine the number of clusters. DBSCAN automatically finds the number of clusters based on data density.
    2. Can find clusters of arbitrary shape: K-Means works best for spherical (round) clusters. DBSCAN can detect clusters of any shape (e.g., elongated, curved).
    3. Handles noise/outliers: K-Means forces every point into some cluster.DBSCAN can label points that don’t belong to any cluster as outliers (noise).
    4. Robust to initialization:K-Means results depend on the initial placement of centroids.DBSCAN does not rely on random initialization, making results more stable.


6. When would you use Silhouette Score in clustering?
   - The Silhouette Score is used in clustering to measure the quality of clusters. It evaluates how well data points are grouped within a cluster compared to other clusters.
   - When to use Silhouette Score:
       - To choose the optimal number of clusters (K):Run clustering with different K values and select the one with the highest silhouette score.
       - To compare different clustering algorithms (e.g., K-Means vs DBSCAN):Helps decide which algorithm gives better-defined clusters.
       - To validate clustering results:Ensures clusters are meaningful and well separated.

7. What are the limitations of Hierarchical Clustering?
  - **Limitations of Hierarchical Clustering:**
  1. Scalability issue:Computationally expensive (O(n²) or more), so it does not work well for very large datasets.
  2. No flexibility after merging/splitting:Once clusters are merged (in agglomerative) or split (in divisive), the decision cannot be undone.
  
  3. Sensitive to noise and outliers:Outliers can create their own clusters or distort the hierarchy.
  4. Choice of distance metric and linkage method matters:Results can vary significantly depending on whether you use single, complete, average, or Ward’s linkage.
  5. Difficulty in determining the optimal number of clusters:Requires analyzing the dendrogram, which can be subjective and unclear for complex data.

8. Why is feature scaling important in clustering algorithms like K-Means?
   - K-Means uses distance-based measures: K-Means relies on Euclidean distance (or similar metrics) to assign points to clusters. If features have different scales, larger-scale features dominate the distance calculation.
   - Prevents bias from large-scale features:Example: Suppose you cluster people based on age (20–60) and income (20,000–100,000).Without scaling, income will overshadow age because of its larger numeric range.
   - Ensures fair contribution of all features: Scaling brings features to a comparable range (e.g., 0–1 or -1 to 1), so each feature influences clustering equally.
   - Improves clustering performance and accuracy: Leads to better-defined clusters and faster convergence in K-Means.

9. How does DBSCAN identify noise points?
   - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) labels a point as noise (outlier) if it does not belong to any cluster based on two parameters:
      - ε (Epsilon): The radius that defines the neighborhood of a point.
      - MinPts: The minimum number of points required within ε to form a dense region (cluster).

10. Define inertia in the context of K-Means.
    - In the context of K-Means clustering, inertia is a measure of how well the data points fit within their assigned clusters. It is defined as the sum of squared distances between each data point and the centroid of the cluster to which it belongs.
    - Low inertia → Data points are close to their centroids → compact and well-defined clusters.
    - High inertia → Data points are far from centroids → poor clustering.




11.  What is the elbow method in K-Means clustering?
     - The Elbow Method is a popular technique to determine the optimal number of clusters (K) in K-Means clustering.
     - In K-Means, increasing the number of clusters always reduces inertia (the sum of squared distances between points and their cluster centroids). However, beyond a certain point, the reduction in inertia becomes marginal, meaning additional clusters do not significantly improve the model.
     - By plotting K (number of clusters) on the x-axis and inertia on the y-axis, we get a curve. Initially, inertia decreases sharply as K increases. After a certain point, the curve forms a distinct “elbow shape".The K value at this elbow point is considered the optimal number of clusters, as it provides the best balance between compact clusters and computational efficiency.

12. Describe the concept of "density" in DBSCAN.
   - In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), the concept of density is the key idea used to form clusters. A cluster is defined as a dense region of points separated by areas of low density.

13. Can hierarchical clustering be used on categorical data?
    - Hierarchical clustering can be applied to categorical data, but it requires some modifications because: Traditional hierarchical clustering relies on distance measures (like Euclidean distance), which are not directly meaningful for categorical variables.

14. What does a negative Silhouette Score indicate?
    - The Silhouette Score measures how well a data point fits into its assigned cluster compared to other clusters. It ranges from -1 to +1. +1 → Perfectly matched to its own cluster, far from other clusters. 0 → On or near the boundary between clusters. -1 (Negative) → Poor clustering → The point is likely assigned to the wrong cluster.
    - Interpretation of Negative Score: A negative score means that the average distance of a point to its own cluster is larger than the average distance to another cluster. In other words, the point is closer to a neighboring cluster than the one it was assigned. This indicates overlapping clusters, poor choice of K, or inappropriate clustering method.

15. Explain the term "linkage criteria" in hierarchical clustering.
    - In hierarchical clustering, the linkage criteria defines how the distance between two clusters is measured when deciding which clusters to merge at each step. Since clusters contain multiple points, we need a rule to calculate the distance between them — this rule is called the linkage criterion.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
    - K-Means clustering works by minimizing the distance of points to their cluster centroids. This approach assumes that clusters are spherical, evenly sized, and equally dense. When clusters have different sizes or densities, K-Means struggles.

17.  What are the core parameters in DBSCAN, and how do they influence clustering?
   - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) relies on two main parameters that control how clusters are formed: ε (Epsilon): Neighborhood Radius and MinPts (Minimum Points): Minimum number of points to form a dense region.
   - How they work together: A core point is one that has at least MinPts neighbors within ε. A border point lies within ε of a core point but has fewer than MinPts neighbors. A noise point is neither a core nor a border point.

18. How does K-Means++ improve upon standard K-Means initialization?
    - Problem with Standard K-Means Initialization: In vanilla K-Means, the initial cluster centroids are chosen randomly. This can lead to: Poor convergence (slow, stuck in local minima), Highly unbalanced clusters,Different results on different runs.
    - K-Means++ Initialization: K-Means++ improves this by choosing initial centroids more intelligently: First centroid is chosen randomly from the dataset.Each subsequent centroid is chosen with probability proportional to the squared distance from the nearest already chosen centroid.This ensures new centroids are far away from existing ones.Repeat until k centroids are chosen. Then, run standard K-Means.
    -  Benefits of K-Means++: Better starting points → clusters are more spread out. Faster convergence → fewer iterations required. More stable results → reduces sensitivity to random initialization. Higher accuracy → often yields a lower within-cluster sum of squares (WCSS).

19. What is agglomerative clustering?
    - Agglomerative Clustering: Agglomerative clustering is a bottom-up hierarchical clustering method. It starts with each data point as its own cluster.Then, in each iteration, it merges the two most similar (closest) clusters based on a linkage criterion (like single, complete, or average linkage). This process continues until all points are merged into a single cluster (or until a stopping condition like a predefined number of clusters is reached).

20.  What makes Silhouette Score a better metric than just inertia for model evaluation?
    - Balances Cohesion & Separation → Not just tight clusters, but also well-separated ones.
    - Scale-independent → Unlike inertia, it doesn’t just keep decreasing with more clusters.
    - Interpretability → A direct sense of cluster quality (-1 to +1).
    - Model Selection → Helps choose optimal k by looking at the maximum Silhouette score.

21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic dataset with 4 centers
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# Scatter plot visualization
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, alpha=0.75, marker='X', label="Centroids")
plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.


```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Store silhouette scores
scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Plot silhouette scores as bar chart
plt.figure(figsize=(6,4))
plt.bar(k_values, scores, color="skyblue")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different k")
plt.show()
```

30.  Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

# Load Iris dataset
iris = load_iris()
X = iris.data

# Perform hierarchical clustering with average linkage
Z = linkage(X, method='average')

# Plot dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, truncate_mode="lastp", p=30, leaf_rotation=90, leaf_font_size=10, show_contracted=True)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample index or (cluster size)")
plt.ylabel("Distance")
plt.show()
```

31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with overlapping clusters
X, y = make_blobs(n_samples=500, centers=3, cluster_std=2.0, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# Plot decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap="viridis")
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap="viridis", edgecolor="k")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c="red", s=200, marker="X", label="Centroids")

plt.title("K-Means Clustering with Decision Boundaries")
plt.legend()
plt.show()
```

34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Store inertia values
inertias = []
k_values = range(2, 7)

# Fit KMeans for K = 2 to 6
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot inertia vs K
plt.figure(figsize=(8,6))
plt.plot(k_values, inertias, marker='o', linestyle='-')
plt.title("K-Means Inertia for K = 2 to 6 (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering?

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data

# Step 2: Apply KMeans (3 clusters for Iris setosa, versicolor, virginica)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Compute Silhouette Coefficients
silhouette_vals = silhouette_samples(X, y_kmeans)
silhouette_avg = silhouette_score(X, y_kmeans)
print(f"Average Silhouette Score: {silhouette_avg:.4f}")

# Step 4: Silhouette plot
n_clusters = 3
y_lower = 10
plt.figure(figsize=(8, 6))

for i in range(n_clusters):
    cluster_silhouette_vals = silhouette_vals[y_kmeans == i]
    cluster_silhouette_vals.sort()
    size_cluster = cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, cluster_silhouette_vals,
                      alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * size_cluster, str(i))
    y_lower = y_upper + 10  # space between clusters

plt.axvline(silhouette_avg, color="red", linestyle="--", label="Average Score")
plt.xlabel("Silhouette Coefficient Values")
plt.ylabel("Cluster")
plt.title("Silhouette Plot for KMeans Clustering on Iris Dataset")
plt.legend()
plt.show()
```

47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count.

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Step 1: Generate noisy blobs
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=42)

# Add random noise points
rng = np.random.RandomState(42)
noise = rng.uniform(low=-10, high=10, size=(50, 2))  # 50 random noise points
X = np.vstack([X, noise])

# Step 2: Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X)

# Step 3: Count clusters and noise
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")
```
