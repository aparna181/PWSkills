#1. What is the role of filters and feature maps in Convolutional Neural Network (CNN)?

Ans:
- A Convolutional Neural Network (CNN) is a specialized type of deep learning model used mainly for image recognition, object detection, and visual data processing.
It automatically learns to extract important features from input images without manual feature engineering.
Two fundamental components that make CNNs powerful are filters (kernels) and feature maps (activation maps).

- Filters:  A filter (also called a kernel) is a small matrix of numbers (weights) that slides over the input image. Each filter performs a mathematical operation called convolution to extract specific patterns from the image.
  - Filters are usually small (e.g., 3×3, 5×5).
  - Each filter learns to detect one specific feature such as:Vertical edges, Horizontal edges, Corners or curves, Color or texture patterns
  - The values inside the filter are learned automatically during training using backpropagation.
  - Example: If a 3×3 filter detects horizontal edges, it will respond strongly to pixel regions that contain horizontal lines in the image.

- Feature Maps (Activation Maps): After the convolution operation, the resulting output is called a feature map or activation map.
This map highlights the areas in the input image where a specific feature (detected by the filter) is present.

  - Characteristics of Feature Maps:
     - Each feature map corresponds to one filter.
     - The brighter (or higher value) areas in a feature map indicate stronger feature detection.
     - Feature maps reduce image dimensions while preserving important information.  

- Filters and feature maps are the core building blocks of CNNs.Filters act as the feature detectors.Feature maps represent where those features exist in the image. Together, they allow CNNs to transform raw pixel data into meaningful patterns that can be used for classification or detection tasks.
Without filters and feature maps, CNNs would not be able to learn or recognize visual patterns effectively.


```
Input Image → [Filter] → Convolution → [Feature Map]
       ↓             ↓
   Raw pixels     Edge detection
```



#2.Explain the concepts of padding and stride in CNNs(Convolutional NeuralNetwork). How do they affect the output dimensions of feature maps?


Ans:
- Padding in CNNs:   Padding is the process of adding extra pixels (usually zeros) around the border of an input image before performing convolution.
Its main purpose is to control the spatial size (height and width) of the output feature map and to preserve important edge information that might otherwise be lost during convolution.

There are two common types of padding:

(a) Valid Padding (No Padding):
- No extra pixels are added around the input.
- The filter only convolves over valid image regions.
- The output feature map becomes smaller than the input.
(b) Same Padding (Zero Padding):
- Zeros are added around the image border.
- Ensures the output feature map has the same spatial dimension as the input (when stride = 1).
- Helps the model learn better edge features since borders are preserved.


Stride in CNNs: Stride defines the number of pixels by which the filter moves over the input image during convolution.
It determines how densely or sparsely the filter scans the image.

Stride = 1:

The filter moves one pixel at a time.

Results in a larger output feature map with high spatial resolution.

More detailed feature extraction but higher computation cost.

Stride > 1:

The filter jumps multiple pixels at a time.

Produces a smaller output feature map.

Reduces computational cost and spatial size but may lose fine details.

```
Effect on Output Dimensions
Parameter	    Increase	       Effect on Output Feature Map
Padding (P)     Increases	           Larger output (preserves spatial dimensions)
Stride (S)	    Increases	           Smaller output (reduces spatial dimensions)
Filter Size (F)	Increases	       Smaller output

```
Padding helps preserve image borders and maintain output size, while stride controls how much the image is downsampled.

#3.Define receptive field in the context of CNNs. Why is it important for deep architectures?


The receptive field in a Convolutional Neural Network (CNN) means the part of the input image that affects one neuron in the output feature map.
In simple words, it shows how much area of the image a neuron is looking at or influenced by.

At the beginning of a CNN, each neuron covers only a small region (like a 3×3 or 5×5 area).
But as the network goes deeper, every neuron starts to depend on larger regions of the original image.
So, in deeper layers, neurons can see a bigger part of the image — this is called a larger receptive field.

Example:

- The first layer may see only small edges or corners (small receptive field).

- The middle layers may see shapes or textures (medium receptive field).

- The last layers may see full objects like faces or cars (large receptive field).

So, as layers increase, the receptive field grows and helps the network understand the whole image better.

Why It Is Important :

- Captures More Information:
A larger receptive field helps the network see more of the image at once, so it can understand the overall structure, not just small parts.

- Better Feature Learning:
Early layers learn simple features (edges), and deeper layers learn complex ones (objects).
This is possible because of the expanding receptive field.

- Context Understanding:
Helps in recognizing an object even if it is big or far away, because the neuron can see a larger area.

- Improves Accuracy in Deep Networks:
In deeper CNNs, a larger receptive field allows better decision-making by combining both local and global information.

The receptive field tells how much of the image a neuron can "see."
It increases with depth and helps CNNs recognize patterns from small details to full objects.
This makes it an important concept for building deep and effective neural networks

#4.Discuss how filter size and stride influence the number of parameters in a CNN.


##Filter Size and Its Influence

A filter (or kernel) is a small matrix of weights used to extract features from an image.
The size of the filter directly affects the number of parameters in a CNN.

Formula for Parameters in One Filter:
Parameters=height of the filter *  width of the filter* number of input channels (e.g., 3 for RGB) +1 for the bias term

Example:

- For a 3×3 filter with 3 input channels:
Parameters = (3×3×3) + 1 = 28 parameters per filter

- For a 5×5 filter with 3 input channels:
Parameters = (5×5×3) + 1 = 76 parameters per filter

Larger filters → more parameters → higher computation and memory cost.

However, smaller filters (like 3×3) are preferred because:

- They capture fine details,

- Require fewer parameters,

- Can be stacked to get the same effect as a large filter with fewer weights.

##Stride and Its Influence

Stride controls how far the filter moves (or “slides”) each time during convolution.

- Stride = 1:
The filter moves one pixel at a time → more overlapping regions → larger output feature map.
This increases the number of activations but does not change the number of parameters in the filter itself.

- Stride > 1:
The filter skips pixels → smaller output feature map → fewer activations.
This reduces the number of total computations, but the number of parameters per filter remains the same.



Stride affects how many times filters are applied, not the number of filter weights.

Filter size directly changes the number of learnable parameters in a CNN — larger filters mean more weights to learn.

Stride does not change the number of parameters but affects how many times filters are applied, influencing the size of the feature map and computational cost.

In practice, small filters (3×3) with stride 1 or 2 are commonly used because they offer a good balance between efficiency and accuracy in deep networks.

#5. Compare and contrast different CNN-based architectures like LeNet AlexNet, and VGG in terms of depth, filter sizes, and performance.

##Architectures
```
| **Architecture** | **Year** | **Developed By**                  | **Used For**                                  |
| ---------------- | -------- | --------------------------------- | --------------------------------------------- |
| **LeNet-5**      | 1998     | Yann LeCun                        | Handwritten digit recognition (MNIST dataset) |
| **AlexNet**      | 2012     | Alex Krizhevsky et al.            | Image classification (ImageNet)               |
| **VGGNet**       | 2014     | Karen Simonyan & Andrew Zisserman | Large-scale image recognition (ImageNet)      |
```
##Comparison in Terms of Depth
```
| **Architecture**    | **No. of Layers (Depth)** | **Description**                                                        |
| ------------------- | ------------------------- | ---------------------------------------------------------------------- |
| **LeNet-5**         | 7 layers                  | Simple architecture: 2 convolutional + 3 fully connected layers        |
| **AlexNet**         | 8 layers                  | 5 convolutional + 3 fully connected layers                             |
| **VGG-16 / VGG-19** | 16 or 19 layers           | Deep network with multiple small convolutional layers stacked together |
```
##Comparison in Terms of Filter Sizes
```
| **Architecture** | **Filter Size**                     | **Stride & Padding**     | **Observation**                                                             |
| ---------------- | ----------------------------------- | ------------------------ | --------------------------------------------------------------------------- |
| **LeNet-5**      | 5×5 filters                         | Stride = 1, no padding   | Detects simple patterns like edges and shapes                               |
| **AlexNet**      | 11×11 (first layer), 5×5, 3×3 later | Stride = 4 initially     | Larger filters for large images                                             |
| **VGG**          | 3×3 filters (all layers)            | Stride = 1, same padding | Uses multiple small filters instead of one large filter for better accuracy |

Smaller filters (3×3) in VGG are more efficient and capture fine details better than large filters.
```
##Comparison in Terms of Performance
```
| **Architecture** | **Dataset Used** | **Top-5 Accuracy** | **Key Achievements**                                                |
| ---------------- | ---------------- | ------------------ | ------------------------------------------------------------------- |
| **LeNet-5**      | MNIST            | ~99%               | Pioneered CNN concept; used for digits recognition                  |
| **AlexNet**      | ImageNet         | ~84.6%             | Introduced ReLU, Dropout, and GPU training                          |
| **VGG-16**       | ImageNet         | ~92.7%             | Demonstrated that deeper networks with small filters perform better |
```

LeNet, AlexNet, and VGG mark important steps in CNN evolution.

- LeNet introduced the basic CNN structure.

- AlexNet expanded it using GPUs, ReLU, and Dropout for large-scale tasks.

- VGG improved performance through depth and small 3×3 filters.

Together, these architectures have shaped the foundation for modern networks like ResNet and Inception, which build upon their ideas for even greater efficiency and accuracy.

#6.Using keras, build and train a simple CNN model on the MNIST dataset from scratch. Include code for module creation, compilation, training, and evaluation.


####Import Libraries

````python
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
import numpy as np
````

### Load the dataset

````python
# use Keras to import pre-shuffled MNIST database
(X_train, y_train), (X_test, y_test) = mnist.load_data()
````

####Visualize


````python
# plot first six training images
fig = plt.figure(figsize=(20,20))
for i in range(10):
    ax = fig.add_subplot(1, 10, i+1, xticks=[], yticks=[])
    ax.imshow(X_train[i], cmap='gray')
    ax.set_title(str(y_train[i]))
````

#### preprocess input image

````python
# rescale to have values within 0 - 1 range [0,255] --> [0,1]
X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255

print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
````

````python
import tensorflow

num_classes = 10
# print first ten (integer-valued) training labels
print('Integer-valued labels:')
print(y_train[:15])

# Check if y_train is already one-hot encoded
if y_train.ndim == 1:  # If y_train is 1D (not one-hot encoded)
    # one-hot encode the labels
    # convert class vectors to binary class matrices
    y_train = tensorflow.keras.utils.to_categorical(y_train, num_classes)
    y_test = tensorflow.keras.utils.to_categorical(y_test, num_classes)
else:
    print("y_train is already one-hot encoded. Skipping to_categorical.")

# print first ten (one-hot) training labels
print('One-hot labels:')
print(y_train[:10])
````

````python
# input image dimensions 28x28 pixel images.
img_rows, img_cols = 28, 28

X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1)

print('input_shape: ', input_shape)
print('x_train shape:', X_train.shape)
````

####model

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, GlobalAveragePooling2D

# build the model object
model = Sequential()

# CONV_1: add CONV layer with RELU activation and depth = 32 kernels
model.add(Conv2D(32, kernel_size=(3, 3), padding='same',activation='relu',input_shape=(28,28,1)))
# POOL_1: downsample the image to choose the best features
model.add(MaxPooling2D(pool_size=(2, 2)))

# CONV_2: here we increase the depth to 64
model.add(Conv2D(64, (3, 3),padding='same', activation='relu'))
# POOL_2: more downsampling
model.add(MaxPooling2D(pool_size=(2, 2)))

# flatten since too many dimensions, we only want a classification output
model.add(Flatten())

# FC_1: fully connected to get all relevant data
model.add(Dense(64, activation='relu'))

# FC_2: output a softmax to squash the matrix into output probabilities for the 10 classes
model.add(Dense(10, activation='softmax'))

model.summary()
````

####Compile the model

````python
# compile the model
model.compile(loss='categorical_crossentropy', optimizer='rmsprop',
              metrics=['accuracy'])
````

####Train the Model

````python
from tensorflow.keras.callbacks import ModelCheckpoint

# train the model
checkpointer = ModelCheckpoint(filepath='model.weights.best.keras', verbose=1,
                               save_best_only=True)
hist = model.fit(X_train, y_train, batch_size=64, epochs=10,
          validation_data=(X_test, y_test), callbacks=[checkpointer],
          verbose=2, shuffle=True)
````

####Evaluation

````python
# load the weights that yielded the best validation accuracy
model.load_weights('model.weights.best.keras')
````

````python
# evaluate test accuracy
score = model.evaluate(X_test, y_test, verbose=0)
accuracy = 100*score[1]

# print test accuracy
print('Test accuracy: %.4f%%' % accuracy)
````

#8.Using PyTorch, write a script to define and train a CNN on the MNIST dataset. Include model definition, data loaders, training loop, and accuracy evaluation.


````python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
````

````python
transform = transforms.Compose([
    transforms.ToTensor(),  # conver to [0,1]
    transforms.Normalize((0.1307,),(0.3081,))
])

train_dataset = datasets.MNIST(root = './data',train = True ,download = True, transform = transform )
test_dataset = datasets.MNIST(root = './data',train = False ,download = True, transform = transform )

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
````

````python
examples = enumerate(train_loader)
batch_idx, (example_data, example_targets) = next(examples)

fig = plt.figure()
for i in range(6):
  plt.subplot(2,3,i+1)
  plt.tight_layout()
  plt.imshow(example_data[i][0], cmap='gray', interpolation='none')
  plt.title("Label: {}".format(example_targets[i]))
  plt.xticks([])
  plt.yticks([])

plt.show()
````

````python
class CNN(nn.Module):
  def __init__(self):
    super(CNN, self).__init__()
    self.conv1 = nn.Conv2d(1,32,kernel_size=3)
    self.conv2 = nn.Conv2d(32,64,kernel_size=3)
    self.dropout1 = nn.Dropout2d(0.25)
    self.dropout2 = nn.Dropout2d(0.5)
    self.fc1 = nn.Linear(9216,128)
    self.fc2 = nn.Linear(128,10)

  def forward(self, x):
    x = F.relu(self. conv1(x))   #[ batch , 32, 26, 26]
    x = F.relu(self.conv2(x))     # [batch, 64, 24, 24]
    x = F.max_pool2d(x,2)              #[batch 64, 12, 12]
    x = self.dropout1(x)
    x = torch.flatten(x,1)
    x = F.relu(self.fc1(x))
    x = self.dropout2(x)
    x = self.fc2(x)
    return F.log_softmax(x, dim=1)

````

````python
from torchsummary import summary

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CNN().to(device)

optimizer = optim.Adam(model.parameters(), lr=0.001)
summary(model,(1,28,28))
````

````python
def train(model, device, train_loader, optimizer, epoch):
  model.train()
  for batch_idx, (data, target) in enumerate(train_loader):
    data, target = data.to(device), target.to(device)


    optimizer.zero_grad()
    output = model(data)
    loss = F.nll_loss(output, target)
    loss.backward()
    optimizer.step()

    if batch_idx % 100 == 0:
      print(f'Train Epoch: {epoch} [{batch_idx * len(data)}/{len(train_loader.dataset)}] Loss: {loss.item():.6f}')
````

````python
def test(model, device, test_loader):
  model.eval()
  test_loss = 0
  correct = 0

  with torch.no_grad():
    for data, target in test_loader:
      data, target = data.to(device), target.to(device)
      output = model(data)
      test_loss += F.nll_loss(output, target, reduction='sum').item()
      pred = output.argmax(dim=1, keepdim=True)
      correct += pred.eq(target).sum().item()

  test_loss /= len(test_loader.dataset)
  accuracy = 100. * correct / len(test_loader.dataset)

  print(f"\nTest set : Average Loss: {test_loss:.4f}, Accuracy: {correct}/{len(test_loader.dataset)}  ({accuracy:.2f}%)\n")
````

````python
for epoch in range(1,6):
  train(model, device, train_loader, optimizer, epoch)
  test(model, device, test_loader)
````

````python
model_path = "mnist_cnn.pth"
torch.save(model.state_dict(), model_path)
print(f"Model saved to {model_path}")
````

````python
loaded_model = CNN().to(device)
loaded_model.load_state_dict(torch.load(model_path))
loaded_model.eval()
print(f"Model loaded sucessfully ")
````

````python
def predict(model, device, data_loader, num_samples=6):
    model.eval()
    data_iter = iter(data_loader)
    images, labels = next(data_iter)
    images, labels = images.to(device), labels.to(device)

    with torch.no_grad():  # disable gradient tracking
        output = model(images)
        preds = torch.argmax(output, dim=1)   # get predicted class labels
        preds = preds.detach().cpu().numpy()  # move to CPU & convert to numpy

    fig = plt.figure(figsize=(8, 6))
    for i in range(num_samples):
        plt.subplot(2, 3, i + 1)
        plt.tight_layout()
        plt.imshow(images[i][0].cpu(), cmap='gray', interpolation='none')
        plt.title(f"Prediction: {preds[i]}, Actual: {labels[i].item()}")
        plt.xticks([])
        plt.yticks([])

    plt.show()
````

````python
predict(loaded_model, device, test_loader)
````

#9.Given a custom image dataset stored in a local directory, write code using Keras ImageDataGenerator to preprocess and train a CNN model.  


````python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import os
import shutil
````

````python
# Step 1: Load tf_flowers dataset
dataset, info = tfds.load('tf_flowers', split='train', with_info=True, as_supervised=True)
num_classes = info.features['label'].num_classes
````

````python
# Step 2: Create folder structure for ImageDataGenerator
base_dir = '/content/flowers_data'
if os.path.exists(base_dir):
    shutil.rmtree(base_dir)
os.makedirs(base_dir, exist_ok=True)

for i in range(num_classes):
    os.makedirs(f'{base_dir}/train/{i}', exist_ok=True)
    os.makedirs(f'{base_dir}/validation/{i}', exist_ok=True)
````

````python
# Step 3: Split dataset into train and validation
train_size = int(0.8 * len(dataset))
train_data = dataset.take(train_size)
val_data = dataset.skip(train_size)
````

````python
# Step 4: Save images into folders
def save_images(ds, subset):
    for i, (image, label) in enumerate(ds):
        label = int(label.numpy())
        img_path = f"{base_dir}/{subset}/{label}/{i}.jpg"
        tf.keras.preprocessing.image.save_img(img_path, image)

save_images(train_data, 'train')
save_images(val_data, 'validation')
````

````python
# Step 5: Create ImageDataGenerators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)
````

````python
# Step 6: Load images from directories
train_generator = train_datagen.flow_from_directory(
    f'{base_dir}/train',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    f'{base_dir}/validation',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)
````

````python
# Step 7: Build CNN model
from tensorflow.keras import layers, models
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])
````

````python
# Step 8: Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
````

````python

# Step 9: Train model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10
)
````

````python
# Plot Training & Validation Accuracy
plt.figure(figsize=(8,5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
````

````python
# Plot Training & Validation Loss
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
````
